```python
#single string classification via Bard
from transformers import pipeline

# Load the zero-shot classification pipeline
classifier = pipeline("zero-shot-classification")

# Define the string
input_string = 'cable f/beam deflection'

# Tokenization
tokens = input_string.split()

# Perform zero-shot classification for each token
results = {}
for token in tokens:
    # Classify using zero-shot classification
    classification = classifier(token, candidate_labels=["manufacturing"])

    # Store the classification score
    results[token] = classification['scores'][0]

# Print the scores for each token
for token, score in results.items():
    print(f"{token}: {score}")
```

```python
#BART zero shot classification
import pandas as pd
from transformers import pipeline
from google.colab import files

# Load the zero-shot classification pipeline
classifier = pipeline("zero-shot-classification")

# Upload Excel file
uploaded = files.upload()

# Read the uploaded Excel file
df = pd.read_excel(next(iter(uploaded.values())))

# Extract the values from the "Spare" column
spare_values = df["Spare"].tolist()

# Initialize dictionary to store the highest scoring token for each spare value
organized_results = {}

# Perform zero-shot classification for each spare value
for spare_value in spare_values:
    # Tokenization
    tokens = spare_value.split()

    # Initialize variables to store the highest scoring token and its score
    max_score = -1
    max_token = ""

    # Perform zero-shot classification for each token
    for token in tokens:
        # Classify using zero-shot classification
        classification = classifier(token, candidate_labels=["manufacturing"])

        # Check if the current token has a higher score than the previous max
        if classification['scores'][0] > max_score:
            max_score = classification['scores'][0]
            max_token = token

    # Store the highest scoring token and its score for this spare value
    organized_results[spare_value] = {"Token": max_token, "Score": max_score}

# Create a DataFrame from the organized results
organized_df = pd.DataFrame(organized_results).T.reset_index().rename(columns={"index": "Spare Value"})

# Save the DataFrame to a new Excel file
output_file = "results.xlsx"
organized_df.to_excel(output_file, index=False)

# Download the output file
files.download(output_file)

print("Organized classification results saved to", output_file)
```

```python
#zero shot classification model form https://huggingface.co/facebook/bart-large-mnli?candidate_labels=K850%2C+Critical%2C+Point%2C+Drier&multi_class=true&text=K850+Critical+Point+Drier
#trained on MultiNLI

from google.colab import files
import pandas as pd
from transformers import pipeline

# Upload the Excel file
uploaded = files.upload()

# Read the Excel file
df = pd.read_excel(next(iter(uploaded)), header=None)  # Assuming the data is in the first sheet and has no header

# Load zero-shot classification pipeline with BART model
classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")

# Define the column to apply zero-shot classification on
column_to_classify = 0  # Assuming the first column, change it accordingly

# Create a list to store the extracted tokens
extracted_tokens = []

# Create a list to store the results
results_list = []

# Perform zero-shot classification for each row in the specified column
for row in df[column_to_classify]:
    input_text = str(row)
    candidate_labels = input_text.split()  # Tokenize input text by space

    # Perform zero-shot classification
    result = classifier(input_text, candidate_labels, multi_label=True)

    # Append the result to the list
    results_list.append(result)

    # Extract the token with the highest score from each result
    max_score_token = None
    max_score = -1
    for label, score in zip(result['labels'], result['scores']):
        if score > max_score:
            max_score = score
            max_score_token = label
    extracted_tokens.append(max_score_token)

# Convert the results list to a DataFrame
results_df = pd.DataFrame(results_list)

# Add a new column for the extracted tokens
results_df['HighPS-Token'] = extracted_tokens

# Save the results to a new Excel file
results_df.to_excel("zero_shot_classification_results.xlsx", index=False)

# Download the Excel file
files.download("zero_shot_classification_results.xlsx")
```

```python
import pandas as pd
from transformers import BartForSequenceClassification, BartTokenizer
from google.colab import files

# Load the BART model and tokenizer
model_name = "facebook/bart-large-mnli"
model = BartForSequenceClassification.from_pretrained(model_name)
tokenizer = BartTokenizer.from_pretrained(model_name)

# Upload Excel file
uploaded_files = files.upload()

# Assuming only one file is uploaded, get its path
file_path = next(iter(uploaded_files))

# Read the uploaded Excel file
df = pd.read_excel(file_path)

# Extract the values from the "Spare" column
spare_values = df["Spare"].tolist()

# Perform classification for each spare value using BART
results = {}
for spare_value in spare_values:
    # Tokenization
    inputs = tokenizer(spare_value, return_tensors="pt", max_length=512, truncation=True)

    # Classification
    outputs = model(**inputs)

    # Get predicted label
    predicted_label = model.config.id2label[outputs.logits.argmax().item()]

    # Store the predicted label
    results[spare_value] = predicted_label

# Add classification results to the DataFrame
df['Classification_Results'] = [results[spare_value] for spare_value in spare_values]

# Save the DataFrame to a new Excel file
output_file = "output_results.xlsx"
df.to_excel(output_file, index=False)

# Download the output file
files.download(output_file)

print("Classification results saved to", output_file)
```

```python
#categorising organisation names
from google.colab import files
import pandas as pd
import re
import spacy

# Load SpaCy English model
nlp = spacy.load("en_core_web_sm")

# Upload Excel file
uploaded_files = files.upload()

# Get the filename
file_name = list(uploaded_files.keys())[0]

# Read the Excel file
df = pd.read_excel(file_name)

# Define replacement dictionary
replacement_dict = {'f/': 'for ', 'w/': 'with ', 'w/o': 'without '}

# Create a new column to store the modified values with initial replacements
df['Modified_Spare'] = df['Spare'].replace(replacement_dict, regex=True)

# Create a new column to store the output with additional replacements
df['Output_Spare'] = df['Modified_Spare'].str.replace(r'(for |with |without )', '', flags=re.IGNORECASE, regex=True)

# Tokenize by space
df['Tokenized_Output'] = df['Output_Spare'].str.split()

# Function to extract organization entities
def extract_org_entities(tokens):
    doc = nlp(" ".join(tokens))
    org_entities = [ent.text for ent in doc.ents if ent.label_ == "ORG"]
    return org_entities

# Apply the function to the Tokenized_Output column
df['Organization_Entities'] = df['Tokenized_Output'].apply(extract_org_entities)

# Save the modified DataFrame to a new Excel file
output_file_name = 'modified_' + file_name
df.to_excel(output_file_name, index=False)

print("Excel file has been successfully modified and saved as:", output_file_name)

# Download the output file
files.download(output_file_name)
```

```python
#final -removed special character upon tokenisation
from google.colab import files
import pandas as pd
import re
import spacy

# Load SpaCy English model
nlp = spacy.load("en_core_web_sm")

# Upload Excel file
uploaded_files = files.upload()

# Get the filename
file_name = list(uploaded_files.keys())[0]

# Read the Excel file
df = pd.read_excel(file_name)

# Define replacement dictionary
replacement_dict = {'f/': 'for ', 'w/': 'with ', 'w/o': 'without ', 'wo/': 'without '}

# Create a new column to store the modified values with initial replacements
df['Input_Spare'] = df['Spare'].replace(replacement_dict, regex=True)

# Create a new column to store the output with additional replacements
df['Output_Spare'] = df['Input_Spare'].str.replace(r'(for |with |without )', '', flags=re.IGNORECASE, regex=True)

# Tokenize by space
df['Tokenized_Output'] = df['Output_Spare'].str.split()

# Function to remove tokens containing special characters
def remove_special_characters(tokens):
    cleaned_tokens = [token for token in tokens if not re.match(r'^\W+$', token)]
    return cleaned_tokens

# Apply the function to the 'Tokenized_Output' column
df['CandidateLabels'] = df['Tokenized_Output'].apply(remove_special_characters)

# Function to extract organization entities
def extract_org_entities(tokens):
    doc = nlp(" ".join(tokens))
    org_entities = [ent.text for ent in doc.ents if ent.label_ == "ORG"]
    return org_entities

# Apply the function to the Tokenized_Output column
df['Organization_Entities'] = df['CandidateLabels'].apply(extract_org_entities)

# Save the modified DataFrame to a new Excel file
output_file_name = 'modified_' + file_name
df.to_excel(output_file_name, index=False)

print("Excel file has been successfully modified and saved as:", output_file_name)

# Download the output file
files.download(output_file_name)
```
